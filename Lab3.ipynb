```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv(".\Life Expectancy Data.csv", header=0)
data.head(10)
```

```python
country_list = data['Country'].unique()
metrics_list = data.keys()
for country in country_list:
    data.loc[data['Country'] == country, metrics_list] = data.loc[data['Country'] == country,metrics_list].interpolate()
data.dropna(inplace=True)
#видалили нульові дані
```

```python
#графіки розподілів факторів (histplot)
fig, axs = plt.subplots(7, 2, figsize=(8, 30))
sb.histplot(data=data.dropna(subset=['Life expectancy ']), x="Life expectancy ", ax=axs[0, 0])
sb.histplot(data=data.dropna(subset=['Adult Mortality']), x="Adult Mortality", ax=axs[0, 1])
sb.histplot(data=data.dropna(subset=['infant deaths']), x="infant deaths", ax=axs[1, 0])
sb.histplot(data=data.dropna(subset=['Alcohol']), x="Alcohol", ax=axs[1, 1])
sb.histplot(data=data.dropna(subset=['Hepatitis B']), x="Hepatitis B", ax=axs[2, 0])
sb.histplot(data=data.dropna(subset=['Measles ']), x="Measles ", ax=axs[2, 1])
sb.histplot(data=data.dropna(subset=['Polio']), x="Polio", ax=axs[3, 0])
sb.histplot(data=data.dropna(subset=['Diphtheria ']), x="Diphtheria ", ax=axs[3, 1])
sb.histplot(data=data.dropna(subset=[' HIV/AIDS']), x= " HIV/AIDS", ax=axs[4, 0])
sb.histplot(data=data.dropna(subset=['GDP']), x="GDP", ax=axs[4, 1])
sb.histplot(data=data.dropna(subset=[' thinness  1-19 years']), x=" thinness  1-19 years", ax=axs[5, 0])
sb.histplot(data=data.dropna(subset=[' thinness 5-9 years']), x=" thinness 5-9 years", ax=axs[5, 1])
sb.histplot(data=data.dropna(subset=['Income composition of resources']), x="Income composition of resources", ax=axs[6, 0])
sb.histplot(data=data.dropna(subset=['Schooling']), x="Schooling", ax=axs[6, 1])
plt.show()
```

```python
Можемо побачити що фактори прогнозування які згідно таблички кореляції мають найбільший прямий 
або зворотній вплив, на практиці не завжди мають такий великий сплив, так "thinness 5-9 years" і
"thinness 1-19 years" мають відносно незначиний вплив, і в той самий час "GDP"
має доволі великий вплив на очікувану тривалість життя хоча цей фактор не мав великого показника 
кореляції
```
